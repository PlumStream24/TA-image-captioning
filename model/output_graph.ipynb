## VGG-LSTM

```python
log_dir = "Oct-26_14-23-30_vgg-lstm\events.out.tfevents.1698305007.Plum-ZephG15.17096.2"
reader = SummaryReader(log_dir)
df = reader.scalars

train_loss = df[df['tag'] == 'Train/Epoch-Loss'].reset_index(drop=True).filter(['step','value'], axis=1).rename(columns= {'step':'epoch', 'value': 'train-loss'})
val_loss = df[df['tag'] == 'Valid/Epoch-Loss'].reset_index(drop=True).filter(['step','value'], axis=1).rename(columns= {'step':'epoch', 'value': 'val-loss'})

loss_df = train_loss
loss_df['val-loss'] = val_loss[['val-loss']]
loss_df

train_bleu = df[df['tag'] == f'Train/BLEU-1'].reset_index(drop=True).filter(['step'], axis=1).rename(columns= {'step':'epoch'})
valid_bleu = df[df['tag'] == f'Valid/BLEU-1'].reset_index(drop=True).filter(['step'], axis=1).rename(columns= {'step':'epoch'})
for i in range(1,5):
    bleu = df[df['tag'] == f'Train/BLEU-{i}'].reset_index(drop=True).filter(['step','value'], axis=1).rename(columns= {'step':'epoch', 'value': f'BLEU-{i}'})
    train_bleu[f'BLEU-{i}'] = bleu[[f'BLEU-{i}']]
for i in range(1,5):
    bleu = df[df['tag'] == f'Valid/BLEU-{i}'].reset_index(drop=True).filter(['step','value'], axis=1).rename(columns= {'step':'epoch', 'value': f'BLEU-{i}'})
    valid_bleu[f'BLEU-{i}'] = bleu[[f'BLEU-{i}']]

display(train_bleu.tail(1))
display(valid_bleu.tail(1))
display(loss_df.tail(1))
train_bleu.plot(x='epoch', y=['BLEU-1','BLEU-2','BLEU-3','BLEU-4'], kind='line')
valid_bleu.plot(x='epoch', y=['BLEU-1','BLEU-2','BLEU-3','BLEU-4'], kind='line')
loss_df.plot(x='epoch', y=['train-loss','val-loss'], kind='line', title='CNN-LSTM Loss')
```
